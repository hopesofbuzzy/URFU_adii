# Введение в `Hugging Face`

Платформа [`Hugging Face`](https://huggingface.co) - [это коллекция](https://habr.com/ru/articles/704592/) готовых современных предварительно обученных (подготовленных) моделей нейронных сетей глубокого обучения (Deep Learning).

Экосистема `HuggingFace` дает возможноть обмениваться подготовленными моделями нейронных сетей или наборами их паратмеров, а также наборами данных в сообществе специалистов.  Такими образом [`Hugging Face`](https://huggingface.co) становится наиболее популярной платформой для размещения современных моделей и библиотек для работы с искусственными нейронными сетями.

Экосистема [`Hugging Face`](https://huggingface.co) включает такие библиотеки, как
* [`transformers`](https://huggingface.co/docs/transformers/index) - основная библиотека для работы с моделями трансформерами;
* [`diffusers`](https://huggingface.co/docs/diffusers/index) - дополнительня библиотека для работы с дифузионными моделями;
* [`datasets`](https://huggingface.co/docs/datasets/en/index) - библиотека для упрощения работы с наборами данных;
* [`hub`](https://huggingface.co/docs/huggingface_hub/) - система версионирования при работы с экосистемой `HF`;
* [`timm`](https://huggingface.co/docs/timm/) - система для работы с классичесики моделями для задач компьютерного зрения (сверточные нейронные сети и нейронные сети - визуальные транасформеры)
* [`peft`](https://huggingface.co/docs/peft/en/index) - специальные методы для дообучения больших моделей;
* [`evaluate`](https://huggingface.co/docs/evaluate/index) - библиотека, содержащая методы оценки результатов работы нейронных сетей.

Отметим, что экосистема `Hugging Face` в основном построена на основе фреймворка `PyTorch`.

[В первую очередь](https://wandb.ai/int_pb/huggingface/reports/An-Introduction-To-HuggingFace-Transformers-for-NLP--VmlldzoyOTgzMjI5) `Hugging Face` был разработан для обрабокти естестественного языка (тексты). ## Быстрый старт

Transformers — [это библиотека](https://matakov.com/hugging-face-transformers-poshagovoe-vvedenie/) от Hugging Face, которая позволяет легко использовать мощные языковые модели вроде BERT, GPT, RoBERTa и многих других. Она содержит как сами модели, так и удобные инструменты для инференса, обучения и обработки текста.

Что такое трансформеры?
Это архитектура нейросетей, которая отлично работает с текстом.
В отличие от старых моделей, трансформеры «понимают» контекст: например, слово «банк» в «речной банк» и «банк денег» — это разные вещи.

Что такое pipeline?
Это готовая «труба», через которую ты пропускаешь данные и получаешь результат.
Пример: pipeline("sentiment-analysis") — модель, которая определяет, позитивный или негативный текст.


Попробуем запустить библиотеку `Transformers`

Для доступа к моделям `Transformers` [есть два способа](https://habr.com/ru/articles/704592/):

* Прямое использование моделей на исходном фреймворке — больше кода, но и больше гибкости.

* Класс `Pipeline` — самый простой способ воспользоваться моделями из `transformers`.

[Если нужно просто попробовать модель](https://huggingface.co/learn/llm-course/ru/chapter2/2) — используем `pipeline`.

Пайплайн в данном случае загружает подготовленный для старта код модели, уже обученной проводить классификации семантической окраски текстов (позитивная или негативная конотация текста).

Пайплайн можно выразить следующим образом
![](https://huggingface.co/datasets/huggingface-course/documentation-images/resolve/main/en/chapter2/full_nlp_pipeline.svg)

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've been ready for a HuggingFace course my whole life.",
        "I hate this so much!",
        "Very strange feelings...",
        "Neutral"
    ]
)
```

Этот код скачает предобученную модель на осное популярной модели для оценки тональности предложений на английском языке `BERT`(`distilbert-base-uncased-finetuned-sst-2-english`) и вернёт метку (`POSITIVE` или `NEGATIVE`) с указанным значением вероятности.

Модель загружается и кэшируется когда вы создадаете объект `classifier`. Если вы перезапустите команду, будет использована кэшированная модель, т.е. загрузки новой модели не произойдет.

В процессе обработки пайплайном текста, который вы ему передали, есть три главных шага:

* Текст предобрабатывается в формат, который понятен модели.
* Предобработанный текст передается в модель.
* Результаты модели проходят через постобработку и вы получаете читаемый ответ.

## Выполнение действий с моделью по шагам

### 1. Предобработка с помощью токенизатора

1. Все модели нейронных сетей (как и все остальные модели машинного обучения) работают с числами.
2. Чтобы подать сырой текст в модель нужно преобразовать его в числа, которые сможет воспринимать модель. Для этого мы используем токенизатор, который будет отвечать за:
  * Разбиение входных данных на слова, подслова или символы (например, пунктуацию), которые называются токенами
  * Сопоставление каждого токена с целым числом
  * Добавление дополнительных входных данных, которые могут быть полезны для модели
3. Вся эта предобработка должна быть выполнена точно так же, как и при предварительном обучении модели, поэтому сначала нам нужно загрузить эту информацию из [`Model Hub`](https://huggingface.co/models). Для этого мы используем класс `AutoTokenizer` и его метод `.from_pretrained()`.
4. Используя имя контрольной точки нашей модели, он автоматически получает данные, ассоциированные с токенизатором модели, и кэширует их (таким образом, они скачиваются только в первый раз, когда вы выполняете приведенный ниже код).



```python

from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

Обратим внимание на то, что такое токен

```python
raw_inputs = [
    "I've been ready for a HuggingFace course my whole life.",
    "I hate this huggingface so much!",
]
tokenizer.tokenize(raw_inputs[1])
```

Обратите внимание на значки `#` в токенах. Так токенайзер выделяет подслова в словах. Это сделано, чтобы не запоминать кучу редких слов и уменьшить хранимый словарный запас. Также это позволяет модели обрабатывать слова, которые она никогда раньше не видела.

Это один из видов токенизации. Всего в трансформерах используются три основных вида токенизации: `Byte-Pair Encoding (BPE)`, `WordPiece`, `SentencePiece`.



Результат представляет собой словарь, содержащий два ключа. Нас интересует
* `input_ids` - содержит две строки целых чисел (по одной на каждое предложение), которые являются уникальными идентификаторами кодов слов в каждом предложении
  
<!-- *  `attention_mask` -  -->



<blockquote>
    
* `input_ids` — массив чисел, каждое из которых соответствует одному токену.

* `attention_mask` — указывает модели, какие токены следует учитывать, а какие игнорировать.

* `token_type_ids` — используется в специальных моделях, в которые подаются пары последовательностей. Например, вопрос-ответ. Тогда в `token_type_ids` эти две последовательности будут обозначены разными метками.
    
</blockquote>

```python
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs['input_ids'])
```

<blockquote>
    
* `padding` — тензоры подаваемые в модель должны иметь одинаковую длину. Если этот параметр = `True`, то коротки последовательности дополняются служебными токенами до длины самой длинной последовательности.

* `truncation` — очень длинные последовательности тоже плохо. Если параметр = `True`, то все последовательности усекаются до максимальной длины. Правила усечения можно найти [тут](https://huggingface.co/docs/transformers/main/en/pad_truncation)

* `max_length` — указываем до скольки токенов усекать последовательность.


</blockquote>

## Загрузка модели

Мы можем загрузить предварительно обученную модель из библиотеки `Transformers` с помощью класса `AutoModel`, у которого также есть метод `from_pretrained()`.


<blockquote>
    

`AutoTokenizer` и `AutoModel*` — универсальные загрузчики, которые подбирают нужный класс модели автоматически.
Существует множество различных архитектур Transformers, каждая из которых предназначена для решения определенной задачи, например:
* *ForMultipleChoice
* *ForQuestionAnswering
* *ForSequenceClassification
* *ForTokenClassification
и другие

Для нашего примера нам понадобится модель с головой классификации последовательности (`ForSequenceClassification` чтобы иметь возможность классифицировать предложения как положительные или отрицательные).
</blockquote>




```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

with torch.inference_mode():
    outputs = model(**inputs)
print(outputs.logits)
```

Наша модель спрогнозировала `[-2.28, 2.3]` для первого предложения и `[ 4.1692, -3.3464]` для второго. Это не вероятности, а логиты, сырые, ненормированные оценки, выведенные последним слоем модели. Чтобы преобразовать их в вероятности, они должны пройти через слой `SoftMax`

```python
with torch.inference_mode():
    predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

Теперь мы видим, что модель спрогнозировала `[0.009, 0.9]` для первого предложения и `[0.9, 0.0005]` для второго. Это узнаваемые оценки вероятности.

Чтобы получить метки, соответствующие каждой позиции, мы можем обратиться к атрибуту `id2label` в конфигурации модели ( более подробно об этом в следующем разделе):



```python
model.config.id2label
```

Теперь мы можем сделать вывод, что модель спрогнозировала следующее:

* Первое предложение: `NEGATIVE: 0.009`, `POSITIVE: 0.9`
* Второе предложение: `NEGATIVE: 0.99`, `POSITIVE: 0.0005`

Мы успешно воспроизвели три этапа конвейера: предобработку с помощью токенизаторов, прохождение входных данных через модель и постобработку! Теперь давайте уделим немного времени тому, чтобы углубиться в каждый из этих этапов.



## Использование `HF` для обработки изображений

Попробуем теперь воспроизвести все указанные действия в одном блоке кода. Код будет позволять проводить классификацию изображений

```python
import torch
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from transformers import AutoImageProcessor, AutoModelForImageClassification

response = requests.get(
    'https://github.com/laxmimerit/dog-cat-full-dataset/blob/master/data/train/cats/cat.10055.jpg?raw=true')
img = Image.open(BytesIO(response.content))

plt.imshow(img); plt.axis(False); plt.show()

img_proc = AutoImageProcessor.from_pretrained(
    'google/vit-base-patch16-224')

model = AutoModelForImageClassification.from_pretrained(
    'google/vit-base-patch16-224')

inputs = img_proc(img, return_tensors='pt')

with torch.no_grad():
    logits = model(**inputs).logits

predicted_id = logits.argmax(-1).item()
predicted_label = model.config.id2label[predicted_id]
print(predicted_id, '-', predicted_label)
```

Тут мы:

* Импортируем два [AutoClass](https://huggingface.co/docs/transformers/main/en/model_doc/auto#transformers.AutoModel)’а: `AutoImageProcessor` и `AutoModelForImageClassification`.
[`AutoClass`](https://huggingface.co/docs/transformers/main/en/model_doc/auto#transformers.AutoModel) (начинается с `Auto`) это специальный класс, который автоматически извлекает архитектуру предварительно обученной модели по ее имени или пути.

* Загружаем картинку по `URL`.

* Загружаем `ImageProcessor`. Это аналог токенайзера, но только для картинок — выравнивает размеры картинок, нормализует и т.д. (об ImageProcessor чуть подробнее ниже). В предыдущем варианте предобработкой занимался сам `Pipeline` где-то в своих недрах. Сейчас же нам придется заниматься этим самостоятельно.

* Загружаем модель. Сама модель представляет собой `PyTorch nn.Module`, который вы можете использовать как обычно при работе с торчом.

* Обрабатываем картинку посредством `ImageProcessor`. `ImageProcessor` возвращает словарь, который подаем на вход модели с оператором распаковки (`**`).

* Все модели `Transformers` возвращают логиты, которые идут перед последней функцией активации (например, `softmax`). Соответственно, нам самим необходимо их обработать, чтобы получить на выходе вероятность или класс.
